**Adda new column ‘Trend’ whose values are based on the 'Daily Return'**

```python
import yfinance as yf
import pandas as pd

# Define the stock symbol
stock_symbol = 'AAPL'

# Get the stock data
stock_data = yf.download(stock_symbol)

# Calculate the daily returns
stock_data['returns'] = stock_data['Close'].pct_change()

# Add a new column 'Trend'
stock_data['Trend'] = 'None'
stock_data.loc[stock_data['returns'] > 0, 'Trend'] = 'Positive'
stock_data.loc[stock_data['returns'] < 0, 'Trend'] = 'Negative'

# Print the stock data with the new column
print(stock_data)
```

Visualize trend frequency through a Pie Chart.

```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt

# Define the stock symbol
stock_symbol = 'AAPL'

# Get the stock data
stock_data = yf.download(stock_symbol)

# Calculate the daily returns
stock_data['returns'] = stock_data['Close'].pct_change()

# Add a new column 'Trend'
stock_data['Trend'] = 'None'
stock_data.loc[stock_data['returns'] > 0, 'Trend'] = 'Positive'
stock_data.loc[stock_data['returns'] < 0, 'Trend'] = 'Negative'

# Create a new dataframe with the frequency of each trend
trend_frequency = stock_data['Trend'].value_counts()

# Plot the trend frequency using a pie chart
trend_frequency.plot.pie(autopct='%.1f%%')
plt.title(f'Trend Frequency for {stock_symbol}')
plt.show()
```

**What was the correlation between the daily returns of different stocks**

```python
import yfinance as yf
import pandas as pd

# Define the stock symbols
stock_symbols = ['AAPL', 'GOOGL', 'MSFT','AMZN']

# Get the stock data
stock_data = yf.download(stock_symbols, group_by='ticker')

# Create a new dataframe with the daily returns
returns_data = pd.DataFrame()
for stock in stock_symbols:
    returns_data[stock] = stock_data[stock]['Close'].pct_change()

# Calculate the correlation between the daily returns
correlation = returns_data.corr()

# Print the correlation
print(correlation)
```

**What was the moving average of various stocks**

```python
import yfinance as yf
import pandas as pd

# Define the stock symbols
stock_symbols = ['AAPL', 'GOOGL', 'MSFT','AMZN']

# Get the stock data from yahoo
stock_data = yf.download(stock_symbols, group_by='ticker')

# Create a new dataframe with the moving average
for stock in stock_symbols:
    stock_data[stock]['moving_average'] = stock_data[stock]['Close'].rolling(window=10).mean()

# Print the moving average data
print(stock_data)
```
